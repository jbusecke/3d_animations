```python
from pyvista_tools.movie import Movie
from pyvista_tools.path import camera_path

import xarray as xr
import numpy as np
ds = xr.open_zarr("gs://leap-persistent/data-library/feedstocks/eNATL_feedstock/eNATL60-BLBT02.zarr", chunks={})
ds = ds.rename({'nav_lon':'lon', 'nav_lat':'lat'})
```

```python
# lets for now just keep the same position and move in closer
nframes = len(ds.time)
nmove = nframes//2
lon = np.repeat(-30, nframes)
lat = np.repeat(25, nframes)
# zoom in then stand still
r = np.hstack([np.linspace(4.0, 2.0, nmove), np.repeat(2.0, nframes-nmove)])
path = camera_path(lon, lat, r)
path
```

```python
factor = 5
test = ds.isel(x=slice(0,None, factor), y=slice(0,None, factor))
test
```

```python
m = Movie(test['vosaline'], camera_path=path, clim=[34, 38], )
```
